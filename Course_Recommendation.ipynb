```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import timeit
import itertools

df = pd.read_csv('iris.csv')
df = df.apply(np.random.permutation,axis=0)
data = df.values
data

x = data[:,:4]
y = data[:,4:]

x_train = x[:120]
x_test = x[120:]

y_train = y[:120]
y_test = y[120:]

class KNN:
    def __init__(self, k=1):
        self.k = k
    
    def train(self,x,y):
        self.x = x
        self.y = y
    
    def predict(self,x):
        label = []
        for i in range(x.shape[0]):
            diff = self.x - x[i]
            diff = diff ** 2
            euc  = np.sum(diff,axis = 1)
            euc  = euc ** 0.5
            indexes = np.argpartition(euc,self.k) #First k minimum distances
            indexes = indexes[0:self.k]
            
            
            l = list(self.y[:,0][indexes])
            dictionary = dict((x,l.count(x)) for x in set(l))
            max_value = max(dictionary,key = dictionary.get)
            label.append(max_value)
        return label
    
    def findBestK(self,validation_x,validation_y,n):
        array = np.zeros(n)
        for i in range(1,n):
            self.k = i;
            array[i] = self.evaluate(validation_y,self.predict(validation_x))
            
        self.k = np.where(array == np.max(array))[0][0]
        #return self.k,np.max(array)
    
    def evaluate(self,y,l):
        count = 0
        for i in range(y.shape[0]):
            if(y[i] == l[i]):
                count += 1
                
        accuracy = (count/len(y))*100
        return accuracy
```

### Test KNN

```python
clf = KNN(3)
clf.train(x_train,y_train)
clf.findBestK(x_test,y_test,10)
labels = clf.predict(x_test)
clf.evaluate(y_test,labels)
```

### Normalize function

```python
def normalize(feature,method = 0): # Normalize
    if(method == 0): #standard Score
        m = feature.mean()
        sd = feature.std()
        feature = (feature - m)/sd
    if(method == 1): #minmax Scaling
        maxi = feature.max()
        mini = feature.min()
        feature = ((feature - mini)/(maxi - mini))
    return feature
```
